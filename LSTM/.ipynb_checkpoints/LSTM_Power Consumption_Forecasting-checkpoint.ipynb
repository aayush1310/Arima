### Splitting Train and Test Data

```python
## Train data (till 2010)

data_train = data.loc[:'2009-12-31',:]['Global_active_power']
```

```python
data_train
```

```python
data_test = data.loc['2010-01-01':,:]['Global_active_power']
data_test
```

```python
print(data_train.shape)
print(data_test.shape)
```

## Prepare Training Data

```python
data_train.head(14)
```

```python
X_train, Y_train = [], []

for i in range(7, len(data_train)-7):
    X_train.append(data_train[i-7:i])
    Y_train.append(data_train[i:i+7])
```

```python
X_train, Y_train = np.array(X_train), np.array(Y_train)
```

```python
X_train.shape, Y_train.shape
```

```python
pd.DataFrame(X_train)
```

```python
pd.DataFrame(Y_train)
```

```python
x_scalar = MinMaxScaler()
X_train = x_scalar.fit_transform(X_train)
```

```python
y_scalar = MinMaxScaler()
Y_train = y_scalar.fit_transform(Y_train)
```

```python
pd.DataFrame(X_train).head()
```

```python
pd.DataFrame(Y_train).head()
```

```python
X_train = X_train.reshape(1098,7,1)
```

## Converted it to 3 Dimensional Data

```python
X_train.shape
```

## Building LSTM Model

```python
model = Sequential()
model.add(LSTM(units=200, activation='relu', input_shape = (7, 1)))
model.add(Dense(7))
```

```python
model.compile(loss='mse', optimizer='adam')
```

```python
model.fit(X_train, Y_train, epochs=100)
```

## Prepare Test Dataset and Test LSTM Model

```python
data_test = np.array(data_test)
```

```python
X_test, Y_test = [], []

for i in range(7, len(data_test)-7):
    X_test.append(data_test[i-7:i])
    Y_test.append(data_test[i:i+7])
```

```python
X_test = x_scalar.fit_transform(X_test)
Y_test = y_scalar.fit_transform(Y_test)
```

```python
X_test.shape
```

```python
X_test = X_test.reshape(331,7,1)
```

```python
X_test.shape
```

```python
Y_pred = model.predict(X_test)
```

```python
Y_pred
```

```python
Y_pred = y_scalar.inverse_transform(Y_pred)
```

```python
Y_pred
```

```python
Y_true = y_scalar.inverse_transform(Y_test)
```

```python
Y_true.shape[1]
```

## Evaluating the Model

```python
def evaluate(y_true, y_predicted):
    scores = []
    #calculate scores for each day
    for i in range(y_true.shape[1]):
        mse = mean_squared_error(y_true[:,i], y_predicted[:,1])
        rmse = np.sqrt(mse)
        scores.append(rmse)
    
    #calculate scores for whole prediction
    total_score = 0
    for row in range(y_true.shape[0]):
        for col in range(y_predicted.shape[1]):
            total_score = total_score+(y_true[row,col]-y_predicted[row,col])**2
    total_score=np.sqrt(total_score/(y_true.shape[0]*y_predicted.shape[1])) 
    
    return total_score, scores
```

```python
evaluate(Y_true, Y_pred)
```
